1. Для чего и в каких случаях полезны различные варианты усреднения для метрик качества классификации: micro, macro, weighted?

micro рассчитывает значение метрик исходя из суммарного числа true positives, 
false negatives, false positives предсказаний.
macro рассчитывает метрики для каждого из классов и вычисляет невзвешенное среднее. Дисбаланс классов не учитывается.
weighted рассчитывает метрики для каждого из клвссов и вычисляет взвешенное среднее с учётом дисбаланса классов.
