1.Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

Лучше максимизировать Precision:
1) Алгоритм должен распознать здоровые семена, отбрать их и не допустить больные.
2) Алгоритм должен распознать пластиковые бутылки для отправки их на переработку как вторсырьё, другой материал попадать не должен, чтобы не сломать технологическую линию.
Лучше максимизировать Recall:
1) Алгоритм должен находить все нарушения в организме человека, далее перепроверяет человек или другой алгоритм.
2) Алгоритм должен рапознать все бракованные изделия, даже если в выборку попадут не бракованные изделия.

2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

F-score - метрика объединяющая в себе точность и полноту, выдавая как результат их гармоническое среднее, а коэффициент  β , позволяет балансирвоать вклад, вносимый точностью, тем самым акцентируя внимание на той метрике, максимизация которой для нас наиболее приоритетна. 
Таким образом, мы используем F-меру, чтобы учесть более ту метрику, которая нам больше важна. Если бы мы просто брали среднее от Precision и Recall, то не могли бы это учесть.
